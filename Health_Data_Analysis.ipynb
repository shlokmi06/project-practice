```python
#Add a Disease column
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Age": [25,30,35,40,45,50],
    "Weight": [65,70,75,80,85,90],
    "BloodPressure": [120,130,135,140,145,150],
    "Cholesterol": [180,190,200,210,220,230],
    "Steps": [8000,7000,6000,5000,4000,3000],
    "HeartRate": [72,75,80,85,88,92]
}

df = pd.DataFrame(data)

# Create Disease column (better condition)
def check_disease(row):
    if (row["BloodPressure"] > 135 or
        row["Cholesterol"] > 200 or
        row["HeartRate"] > 85):
        return 1   # Disease
    else:
        return 0   # No Disease

df["Disease"] = df.apply(check_disease, axis=1)

print(df)
```

```python
#plotting Age vs Disease
plt.scatter(df["Age"], df["Disease"], color='red')

plt.xlabel("Age")
plt.ylabel("Disease (0 = No, 1 = Yes)")
plt.title("Age vs Disease Relationship")

plt.show()
```

```python
# Basic analysis
print("Average Values:\n")
print(df.mean())

print("\nMaximum Values:\n")
print(df.max())

print("\nMinimum Values:\n")
print(df.min())
```
